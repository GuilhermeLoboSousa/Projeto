```python
data = []  # Lista para armazenar os dados extrados

with open('AMP_EXP1.txt', 'r') as arquivo:
    for linha in arquivo:
        # Separar os campos da linha
        campos = linha.strip().split('\t')

        if len(campos) >= 8: #algo so pq sim
            uniprot_id = campos[2]
            sequence = campos[7]

            # Adicionar as informaes extradas  lista de dados
            data.append((uniprot_id, sequence))
data
```

```python
from propy import PyPro
from propy.GetProteinFromUniprot import GetProteinSequence
from Bio.SeqUtils.ProtParam import ProteinAnalysis
import urllib.error

from urllib.parse import quote

all_descriptors = []
proximos=[]

for tupple in data:
    try:
        protein_id = quote(tupple[0])
        protein_sequence = GetProteinSequence(protein_id)
        DesObject = PyPro.GetProDes(protein_sequence)
        
        try:
            X = ProteinAnalysis(tupple[1])
            descriptors = [
                protein_id,
                1,
                tupple[1],
                X.length,
                X.aromaticity(),
                X.isoelectric_point(),
                DesObject.GetAAComp(),
                DesObject.GetDPComp(),
                DesObject.GetCTD()
            ]
            try:
                descriptors.insert(4, X.molecular_weight())
            except ValueError as e:
                print(f"Ignorando erro de valor para o clculo de peso molecular da tupla {tupple}: {e}")
            
            all_descriptors.append(descriptors)
        except AttributeError as e:
            print(f"Ignorando erro de atributo para a tupla {tupple}: {e}")
    except (TimeoutError, urllib.error.URLError) as e:
        proximos.append(tupple)
        print(f"Erro ao obter sequncia de protena para a tupla {tupple}: {e}")

# Exibir a lista de descritores para todas as tuplas

#for descriptors in all_descriptors:
    #print(descriptors)


# Exibir a lista de descritores para todas as tuplas

#for descriptors in all_descriptors:
    #print(descriptors)

```

```python
def extract_values(data):
    values = []
    if isinstance(data, list):
        for item in data:
            values.extend(extract_values(item))
    elif isinstance(data, dict):
        for value in data.values():
            values.extend(extract_values(value))
    else:
        values.append(data)
    return values



nova_data=[]
for x in teste:

    values = extract_values(x)
    nova_data.append(values)
    print(values)


```
