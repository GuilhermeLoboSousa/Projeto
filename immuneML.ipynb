Method for caling Motif script and counting the frequencies of motif

```python
def call_and_parse_motif_on_train (x_data) :
    if os.path.exists("pos_train_seq_new.fasta"):
        os.remove("pos_train_seq_new.fasta")#remover se exisitir esse ficheiro
    if os.path.exists("neg_train_seq_new.fasta"):
        os.remove("neg_train_seq_new.fasta")
    if os.path.exists("motif_output.occurrences"):
        os.remove("motif_output.occurrences")
        #os.rename('C:\\Users\\bipas\motif_output_Sept14.occurrences', 
        #          'C:\\Users\\bipas\motif_output_Sept14_'+str(n)+'.occurrences')
    fp= open("pos_train_seq_new.fasta","a+") # abre como escrita e leitura , i guess, guardado em fp
    fn= open("neg_train_seq_new.fasta","a+")
    for rows in x_data:
        if(rows.item(1) == 1):#junta no positivo
            fp.write('>')
            fp.write(str(rows.item(0))) # adicoona o identificador > nome do pepetido ou sp...
            fp.write('\n')
            fp.write((str(rows.item(2))) + '\n')# paragrafo e junta a sequencia
            #fp.write('\n')
    for rows in x_data:
        if(rows.item(1) == 0): #junta no negativo 
            fn.write('>')
            fn.write(rows.item(0)) #identificador
            fn.write('\n')
            fn.write((str(rows.item(2)) )+ '\n') #sequencia do pepetido
            #fn.write('\n') 
    fp.close()  
    fn.close()
    #print("running motif script")
    pl_script = subprocess.Popen(['C:\Strawberry\perl\\bin\perl.exe', 
                                  'C:\\Users\\bipas\MERCI\MERCI-U.pl', 
                                  '-p', "C:\\Users\\bipas\\pos_train_seq_new.fasta", 
                                  '-n', "C:\\Users\\bipas\\neg_train_seq_new.fasta",  
                                  '-o', "motif_output",
                                  #'-c', "BETTS-RUSSELL",
                                  '-k', "ALL"
                                 ]) #utilização do perl previamente instalado
                                  #'-c', "BETTS-RUSSELL"])
    pl_script.communicate()
    #print("motif script end, now parsing")
    occ_file='C:\\Users\\bipas\motif_output.occurrences' #nome do arquivo de saida
    f = open(occ_file, "r") #abrir o arquivo em modo leitura
    occ_word_list = f.read().split()
    f.close()
#procura por padroes no script de entrada, guarda esses resultados num ficheiro out, lê esse ficheiro para extrair as informações mais relevantes
    #print (x_data.shape)
    Xm=x_data[:,3:] # os dados a partir da 4 coluna (SeqL)
    #print (Xm.shape)
    N=x_data[:,0].shape #algo que fica tuple(nºlinhas,)
    b=np.zeros((N[0],1)) # cria matriz zero com o mesmo numero de linhas , genero de uma coluna toda em zero
    Xm = np.hstack((b, Xm))#adicionar esta nova matriz de zero aos dados , ou seja no array o primeiro elemeto passa a ser 0.0
    i=0
    max = 0
    for x in x_data: #vai a cada array
        name='>'+x[0] #vai ter so o identificador do peptido
        #print(x[0])
        m=occ_word_list.count(name)
        #print(m)
        Xm[i,0]=m #substituir o vetor de zero pelo motif???
        #if (m>max):
        #    max = m
        i=i+1
    return Xm
```
